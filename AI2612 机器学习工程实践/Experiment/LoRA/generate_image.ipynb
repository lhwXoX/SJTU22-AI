```python
from diffusers import AutoPipelineForText2Image
import torch
import os
# os.environ['CUDA_VISIBLE_DEVICES'] = '3'
pipeline = AutoPipelineForText2Image.from_pretrained("./stable-diffusion", torch_dtype=torch.float16).to("cuda:3")
pipeline.load_lora_weights("./output/comic/checkpoint-500", weight_name="pytorch_lora_weights.safetensors")
prompt = "A dog in the style of comic"
image = pipeline(prompt).images[0]
image
```
